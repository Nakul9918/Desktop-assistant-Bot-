```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('insurance.csv')
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df = pd.get_dummies(df,columns =['sex','smoker','region'],drop_first = True)
```

```python
plt.figure(figsize = (10,9))
sns.heatmap(df.corr(),annot = True,cmap = 'coolwarm')
plt.title("Feature correlation Heatmap")
plt.show()
```

```python
x = df[['age','bmi','smoker_yes']]
y = df['expenses']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)
pred_rf = rf.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

def evaluation(name, y_test,y_pred):
  mae = mean_absolute_error(y_test,y_pred)
  mse = mean_squared_error(y_test,y_pred)
  rmse = np.sqrt(mse)
  print(f"{name} - MAE: {mae:.2f}, RMSE: {rmse:.2f}")

evaluation("Random Forest",y_test,pred_rf)
```

```python
import pandas as pd

# New person's data
new_input = pd.DataFrame([{
    'age': 18,
    'bmi': 33.8,
    'smoker_yes': 0  # Non-smoker
}])
```

```python
predicted_expenses = rf.predict(new_input)
print(f"Predicted expenses for the new person: {predicted_expenses[0]:.2f}")
```

```python
actual = 1725.55
predicted = predicted_expenses[0]

accuracy = 100-(abs(actual-predicted)/actual*100)
print("prediction accuracy",round(accuracy,2),"%")
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("Mean Absolute Error (MAE): ₹", round(mae, 2))
print("Root Mean Squared Error (RMSE): ₹", round(rmse, 2))
```

```python
new_input = pd.DataFrame([{
    'age':28,
    'bmi':36.4,
    'smoker_yes':1

}])

predicted_charge = rf.predict(new_input)
print("priedicted insurance premium:",round(predicted_charge[0],2))
```

```python
actual = 51194.56
predicted = predicted_charge[0]

accuracy = 100 - (abs(predicted - actual) / actual * 100)
print("Prediction Accuracy: ", round(accuracy, 2), "%")
```
